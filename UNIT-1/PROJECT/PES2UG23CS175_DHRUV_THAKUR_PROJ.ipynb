```python
from transformers import pipeline
ner = pipeline("ner", aggregation_strategy="simple")
```

```python
resume_text = """
Dhruv Thakur
BTech in Computer Science, PES University
Intern at Council of Scientific and Industrial Research (CSIR)
Worked on OpenCV, TensorFlow, NLP, Computer Vision.
"""
```

```python
entities = ner(resume_text)

for e in entities:
    print(e["entity_group"], ":", e["word"])
```

```python
name = None
universities = []
companies = []

for e in entities:
    if e["entity_group"] == "PER" and name is None:
        name = e["word"]
    elif e["entity_group"] == "ORG":
        if "university" in e["word"].lower() or "college" in e["word"].lower():
            universities.append(e["word"])
        else:
            companies.append(e["word"])

print("Name:", name)
print("University:", universities)
print("Company:", companies)
```

```python
result = {
    "Name": name,
    "University": universities[0] if universities else "Not Found",
    "Company": companies[0] if companies else "Not Found"
}

result
```
